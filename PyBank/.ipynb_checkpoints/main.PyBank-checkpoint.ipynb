````python
from pathlib import Path
import csv

import pandas as pd
from pathlib import Path
````

````python
budget_data_path = Path('budget_data.csv')

budget_data = pd.read_csv(budget_data_path)


budget_data.head()
````

````python
total = budget_data.loc[:, 'Profit/Losses'].sum()
````

````python
total_months = budget_data.loc[:,'Date'].count()
````

````python
greatest_increase = budget_data.loc[:, ['Date', 'Profit/Losses']].max()
````

````python
greatest_decrease = budget_data.loc[:, ['Date', 'Profit/Losses']].min()
````

#------ summary-------

````python
print(f" Total : {total}")
print(f"Total month: {total_months}")
print(f"Greatest increase in profits: {greatest_increase}")
print(f"Greatest decrease in profits: {greatest_decrease}")
````
